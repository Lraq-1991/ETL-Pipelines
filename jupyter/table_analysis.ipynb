```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import psycopg2 as psy
```

```python
def fetch_data(pg_conn, query):

    try:
        # Database connection
        connection = psy.connect(
            database = pg_conn['database'],
            user = pg_conn['user'],
            host = pg_conn['host'],
            password = pg_conn['password'],
            port = pg_conn['port']
        )

        # Open a cursor to perform database operations
        cursor = connection.cursor()

        # Execute command
        cursor.execute(query)

        rows = cursor.fetchall()

        # Make the changes persistents
        connection.commit()

        # Close cursor and connection
        cursor.close()
        connection.close()

        # Transform to DataFrame
        df = pd.DataFrame(rows)
        
    except Exception as err:
        print(err)

    return df
```

```python
pg_conn = {
    'database': 'customers',
    'user': 'airflow',
    'host': '172.18.0.4',
    'password': 'airflow',
    'port': 5432
}

query = 'SELECT * FROM customers LIMIT 100'

df = fetch_data(pg_conn, query)

headers = ['customerid','surname','creditscore','geography','gender','age','tenure','balance','numofproducts','hascrcard','isactivemember','salary','exited']

df = df.set_axis(headers, axis = 1)
```
